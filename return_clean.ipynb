```python
import pandas as pd
import govt_func as gf
import datetime
from imp import reload

import numpy as np
from dateutil.relativedelta import *
```

# Import Data

```python
provider_uri = "data/avg_stock_price.xlsx"

df_pre = pd.read_excel(provider_uri, sheet_name="Sheet1")
df_close = pd.read_excel(provider_uri,sheet_name="Sheet2")
df_pre = df_pre.replace(0,np.nan)
df_close = df_close.replace(0,np.nan)
df_pre
```

# Clean monthly data

```python
df_pre = df_pre.rename(columns = {"Unnamed: 0":"DATE"})
col = []
for item in df_pre["DATE"]:
    i = datetime.datetime.strptime(item, '%Y-%m-%d')
    col.append(i)
df_pre["DATE"] = col
df_pre
```

```python
df1 = df_pre.set_index("DATE")
df1
```

```python
df_close = df_close.rename(columns = {"Unnamed: 0":"DATE"})
df_close
```

```python
df2 = df_close.set_index("DATE")
df2
```

# Get Monthly Return

```python
# merge df1 df2
df = pd.concat([df1,df2]).sort_index()
df
```

```python
df_return = df.shift(-1)
df_return = df_return / df - 1
df_return
```

```python
df_return = df_return.iloc[::2]
df_return
```

```python
df_return = df_return.stack(dropna= False)
df_return
```

```python
df = df_return.reset_index()
df = df.rename(columns={"level_1":"SECU_CODE", 0: "RETURN"})
df = df.set_index("DATE")
df
```

# Write File

```python
df.to_csv("cleaned_data/cleaned_return.csv")
```
