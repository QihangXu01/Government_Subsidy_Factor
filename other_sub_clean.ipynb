```python
import pandas as pd
import govt_func as gf
import datetime
from imp import reload

import numpy as np
from dateutil.relativedelta import *
```

# Import Data

```python
other_uri = "data/other_income.xlsx"
df_other = pd.read_excel(other_uri)
profit_uri = "data/profit.xlsx"
df_profit = pd.read_excel(profit_uri)
df_other = df_other.drop("证券名称", axis=1)
df_profit = df_profit.drop("证券名称", axis=1)
df_other = df_other.rename(columns={"CITICS_INDUSTRY": "INDUSTRY"})
df_profit = df_profit.rename(columns={"CITICS_INDUSTRY": "INDUSTRY"})
df_other = df_other.dropna(axis=0, subset="INDUSTRY")
df_profit = df_profit.dropna(axis=0, subset="INDUSTRY")
df_other = df_other.drop(
    df_other.iloc[:, 2:].filter(like="Y", axis=1).columns, axis=1
)  # exclude annual statement
df_profit = df_profit.drop(
    df_profit.iloc[:, 2:].filter(like="Y", axis=1).columns, axis=1
)  # exclude annual statement
df_other
```

# Get other income in Quarter 

```python
df = df_other.set_index(["SECU_CODE","INDUSTRY"])
df_other = gf.get_quarter(df)
df_other
```

# Get Sum of Other Income by Industry

```python
df_ind_sum = df_other.groupby("INDUSTRY").sum()
df_ind_sum = df_ind_sum.stack().reset_index()
df_ind_sum = df_ind_sum.rename(columns={"level_1": "QUARTER", 0: "SUM_OTHER"})
df_ind_sum = df_ind_sum.set_index(["INDUSTRY", "QUARTER"])
df_ind_sum
```

# $Support_i$ = $other_i$/sum($subsidy_s$)

```python
df = df_other.stack()
df = df.reset_index()
df = df.rename(columns={"level_2": "QUARTER", 0: "OTHER_SUB"})
df
```

```python
df = df.set_index(["INDUSTRY", "QUARTER"])  # synchronize index
df
```

```python
df_other = pd.merge(
    left=df, right=df_ind_sum, left_index=True, right_index=True
)  # merge the Dataframe to make other and industry sum in the same Dataframe
df_other = df_other.reset_index()
df_other = df_other.sort_values(["QUARTER", "SECU_CODE"])
df_other["F_SUPPORT"] = df_other["OTHER_SUB"] / df_other["SUM_OTHER"] # get factor value
df_other = df_other.set_index(["QUARTER", "SECU_CODE", "INDUSTRY"])
df_other
```

# Get $profit$ in quarter

```python
df = df_profit.set_index(["SECU_CODE", "INDUSTRY"])
df_profit = gf.get_quarter(df)
df_profit
```

```python
df = df_profit.stack()
df = df.reset_index()
df = df.rename(columns={"level_2": "QUARTER", 0: "PROFIT"})
df = df.set_index(["QUARTER", "SECU_CODE", "INDUSTRY"])  # synchronize index
df
```

```python
df_other = pd.merge(
    left=df_other, right=df, left_index=True, right_index=True
)  # merge the dataframe to make other and profit in the same dataframe
df_other
```

# dependence = subsidy/profit

```python
df_other["DEPENDENCE"] = df_other["OTHER_SUB"] / df_other["PROFIT"]
# subsidy > profit : dependence = 1
# other: original value
df_other["F_DEPENDENCE"] = df_other.DEPENDENCE
df_other["F_DEPENDENCE"] = df_other.F_DEPENDENCE.mask(
    (df_other.OTHER_SUB > df_other.PROFIT), 1
)
df_other
```

```python
# delete redundant columns
df_other = df_other.drop(
    ["OTHER_SUB", "SUM_OTHER", "PROFIT", "DEPENDENCE"], axis=1
)
df_other
```

# Rank F_SUPPORT

```python
df_other["F_SUPPORT"] = df_other.groupby(by="QUARTER").rank(axis=0)["F_SUPPORT"]
df_other
```

# Standardization

```python
df = df_other
mean = df.groupby(by="QUARTER").mean()
std = df.groupby(by="QUARTER").std()

df = gf.get_standard(df, mean, std)
df
```

```python
df_other = df.drop(["F_SUPPORT_mean", "F_DEPENDENCE_mean", "F_SUPPORT_std", "F_DEPENDENCE_std"],axis = 1)
df_other
```

# Exclude Outlier

```python
df_other[df_other.F_SUPPORT > 3] = np.nan
df_other[df_other.F_SUPPORT < -3] = np.nan
df_other[df_other.F_DEPENDENCE > 3] = np.nan
df_other[df_other.F_DEPENDENCE < -3] = np.nan
df_other
```

# F_OTHER = F_SUPPORT - F_DEPENDENCE

```python
df_other["F_OTHER"] = df_other["F_SUPPORT"] - df_other["F_DEPENDENCE"]
df_other
```

```python
df_other = df_other.reset_index().sort_values(by = ["QUARTER", "SECU_CODE"])
df_other = df_other.set_index("QUARTER")

df_other
```

# Write File

```python
df_other.to_csv("cleaned_data/cleaned_other.csv")
```
